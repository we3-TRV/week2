```python
from nltk.tokenize import RegexpTokenizer
from nltk.probability import FreqDist
!pip install pdfminer.six
from pdfminer.high_level import extract_text
```

```python
#extract the text from pdf file
text=extract_text('/content/2010.00462.pdf')

print(text[:20])
```

```python
#create an instance of tokenizer using NLTK RegexpTokenizer
tokenizer=RegexpTokenizer(r'\w+')

#Tokenize the text read from PDF File
tokens=tokenizer.tokenize(text)

print(tokens)
```

```python
#find the frequency distribution
freqdist=FreqDist(tokens)

print(freqdist)
```

```python
#find words whose length is greater then 5 and frequency is greater then 20
long_freq_words=[words for words in tokens if len(words)>5 and freqdist[words]>7]

print(long_freq_words)
```

```python
#Represent in Graph
FreqDist(long_freq_words).plot()
```
